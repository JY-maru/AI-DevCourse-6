## chatGPT

- Transformer 
    - 초 거대 모델의 초석 : 거대 언어 모델(LLM) 기반 기술
    - Google 2017년 공개
    - 순차 정보처리 특화 (RNN -> LSTM -> GRU -> Transformer)
    - 핵심기술 : `self-attention`, `multi-head attention`
    - 주요 학습 방법 : Pretraining(self-supervised Learing) -> fine tuning(downstream task)

- chatGPT
    - GPT : Generative Pre-trained Transformer
    - Transformer 기반 자연어 생성 모델

- 활용 사례 (보조적 수단 증가)
    - 논문 작업 
    - 대본 작성
    - 시험 질의 응답
    - 프로그래밍 작성(수정 및 교정)

- 개발 시 특정 언어에 대한 공부보단 문제 정의에 커버할 수 있는 공부를 하자.
- 저급 프로그래머는 활용되지 않음.
