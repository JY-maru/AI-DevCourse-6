# BeautifulSoup
- HTML paser


```python
import requests

res = requests.get('http://www.example.com')
print(res.text)
```

- BeautifulSoup 불러오기

```python
from bs4 import BeautifulSoup


soup = BeautifulSoup(res.text, "html.parser")
```

- indent 적용


```python
# 보기 편한 형태로 변환
print(soup.prettify())
```

- title 가져오기

```python
soup.title
```

- head 가져오기

```python
soup.head
```

- body 가져오기

```python
soup.body
```

```python
# <h1> 태그로 감싸진 요소 찾기
h1 = soup.find('h1')
h1
```

- 태그 모두 가져오기

```python
# <p>로 감싸진 요소 찾기
soup.find_all('p') # list
```
