## 1. 포트폴리오란 무엇인가?
- Port + Folio : 묶음들을 들고다니는 것!
- Portfolio는 자신의 기술, 능력, 그리고 경험을 시각화한 일종의 경력증명서
- 과제, 업무, 프로젝트, 기타 활동에 대한 소개 및 자신의 능력을 입증할 수 있는 자료
- 가지고다니다(port) + 여러 잎사귀(Folio)를 합친 말로 본래는 서류가방이나 서류집을 의미한다고 한다.


### 꼭 작성해야 하나요?
- 자기 커리어 로드맥의 전반을 효과적으로 보여줄 수 있고, 역으로 자기 점검을 하고 이후의 커리어 전반을 관리할 수 있는 도구로 활용할 수 있어 매우 중요하다고 할 수 있다. 

### 포트폴리오의 목적
- 포트폴리오는 자신의 업무 능력과 전문성을 보여주기 위함.
- 업무 수행 능력과 문제 해결 능력을 보여줄 수 있는 최적의 수단.

### 포트폴리오의 유형
- PPT, Word PDF형식의 전자 포트폴리오
- 웹사이트 형식
- 블로그 형식
- Git repository
- Notion을 활용한 형식



### Good Case

1. 여러 지원자 중 돋보이게 하는 특별한 수단.
2. 이력서보다 더 본인의 실력과 경험을 구체적이며 생생하게 전달.
3. 지원자 본인에 대한 첫번째 혹은 유일한 인상을 결정
4. 그체적인 경험과 성과를 시각적으로 전달.

### 기초작업
1. 그동안 모든 경험과 활동, 업무들을 나열
- 프로젝트 제목, 수핻ㅇ기간
- 프로젝트 목표와 나의 역할
- 프로젝트에서 사용한 기술, 툴 (도커와 같은 기술스텍, ai모델링의 경우 사용한 아키텍처나 방법론)
- 프로젝트 업무 수행 프로세스
- 업무 수행 중 마주쳤던 어려움 또는 해결해야 했던 문제, 이슈
- 업무 수행을 통해 얻어낸 결과 또는 성과 $\to$ 정량적으로 표현할 수록 좋다. (성능 15% 증가보다는 기준이 되는 정량을 제시. ex. 쓰루풋 14%증가)
- (오픈 프로젝트/ 과제 등 공개 가능한 경우) 관련 링크 첨부하기.

2. 포트폴리오를 지원 혹은 제출하려는 대상 청중이 누구인지에 따라 경험과 활동, 업무 선별

3. 선별 내용을 청중이 선호하는 방식으로 재구성


### 활용 가능 툴

1. 웹사이트 폼
ex. 프로그래머스 이력서 폼

2. PPTX

3. 웹사이트
- 자기를 소개할 웹사이트를 직접 만들기 (Ex. Wordpress, Google Site등 무료 활용 가능)
- 실시간 데모나 Interactive한 내용을 담아야할 때 효과적

4. 기술 블로그
- 자신의 생각, 아이디어나 지식을 정리/공유하기 위함.
- 인터뷰어가 자신의 블로그를 보게 되기도 함.
- ai분야에서 Paper Review에 대한 기술 블로그 운영하기도 함.

5. Github repository
- READMD.md로 포트폴리오로 활용
- 프로젝트 관련 해서 한번에 어필 가능. 개발로그(Commit, PR, Issue)에 대한 어필도 가능하다는 장점.

### 공통사항
- 공개 포트폴리오 유형은 **지속적 관리와 업데이트** 필요
- 시각적 화려도 중요하지만, 얼마나 본인의 서사나 장점을 효과적으로 보여줄 수 있을지를 고민.
- 지원하는 회사나 업종에 맞게 적합한 포트폴리오를 선택하기.

### (1) 명확한 목적 설정
- 회사, 기관, 연구실에 해당하는 타깃의 문화나 요구사항 파악하기
- 이에 대한 스스로의 장점과 능력 반영하기
- 실제로 수행하지 않은 업무나 분야라도 그를 위한 관심이나 노력이 반영되어야 함. (ex. 논문 리뷰해오는 형식의 면접도 존재. 논문과 관련된 질의응답 시, 포폴에도 포함되어 있으면 좋다. )

- 모든 능력을 표현하려고 하기보다는 다른 사람과 비교할 수 있는 특별한 고유 경험이나 스킬을 작성하기.
(당연한 능력 (Ex.개발자로서의 성실함, 과제 이행, 발표 대회 1등 등..)을 길게 서술하지 않는다. 하나를 쓰더라도 **딥하고 자세하게!** )
- 특정 포인트의 감성 극대화하기. (데이터 몇만개 하드코딩,...)
- 극복 사례가 어떻게 성과로 이어졌는지 표현하기.

- 부캠 과정을 수간한 사람은 많고 유사한 부분. 그치만 이 과정 자체가 자신의 강점이나 특별한 능력을 발휘하도록 어떤 도움이 됐는지 표현한다면 좋은 경험 사례로 인정됨.

- 검은색 화면의 포폴은 추천 x.

### (2) 프로젝트의 주요 성과 강조
- 자신이 참여한 프로젝트에서 자신이  만들어낸 성과를 구체적인 숫자나 데이터를 통해 보여주는 것이 효과적
- 단순 수치나 데이터가 아닌 청중으로 하여금 성과가 어느정도 효과가 있었을지 상상할 수 있는 방식으로 나타내야함.
(ex. 물이 월드컵경기장 4개를 가득 채울 양)
- 정상적인 부분에서도 정량적으로 표현 가능한 영역이 많으며 그런 부분들이 실제 프로젝트의 KPI로 정의되어. 프로젝트를 기획하고 진행할 때도 이런 부분들을 고려하게된다.
- 결국 성과를 잘 내는 사람을 원하기 때문에 청중이 들었을 때 좋아할만한 성과들을 중심으로 포트폴리오를 구성할 것.

- 수치 : 상대가 의미있게 느낄만한 수치로 변환해서 표현하기.
- 프로젝트 기획 시, 포폴에 어떻게 기입할지를 생각하고 기획하면 좋다.

### (3) 복잡한 내용은 간단하고 명료히 작성
- 내용이 길어지고 늘어질 수 있다.
- 처음 포트폴리오를 보게되는 사람은 지원자의 전문성 자체를 평가할 수 있는 사람이 아닐 확률이 높아 이를 고려해서 최신 기술을 이해하기 쉽게 풀어쓰는 능력 자체가 중요.
- 이를 통해 자신이 수행한 업무를 충분히 잘 알고있는지 청중은 앎.

### (4) 종료된 프로젝트에도 지속적 업데이트 필요
- 과거의 기술에 대한 보완으로 현재 더 나은 기술을 적용해서 새로운 해결방식을 제시할 수 있음.
- 지속적인 업데이트를 통해 포폴 업그레이드 하기.
