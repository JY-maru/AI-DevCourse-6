# 확률 이론
## 확률 변수
확률 변수 $X$는 표본의 집합 $S$의 원소 $e$를 실수값 $X(e)=x$에 대응시키는 함수.
- 대문자 $X, Y, \cdots$ : 확률변수
- 소문자 $x,y, \cdots$: 확률 변수가 가질 수 있는 값.
-확률 $P$는 집합 $S$의 부분집합을 실수값에 대응시키는 함수. 
    - $P(X=x)$
    - $P(X≤x)$

## 확률 변수의 성질 
- 덧셈법칙
> $$P(X) = \sum_Yp(X,Y)$$
- 곱셈법칙
> $$P(X, Y) = p(X|Y)p(Y) = p(Y|X)p(X)$$
- 베이즈 확률
> $$\begin{aligned}p(Y|X) =& \frac{p(X|Y)p(Y)}{\sum_Yp(X|Y)p(Y)} \\
posterior =& \frac{likelihood\times priors}{normalization}
\end{aligned}$$
>- posterior : 사후확률
>- likelihood : 가능도(우도)
>- prior : 사전확률
>- normalization : $Y$와 상관없는 상수, $X$의 경계확률(marginal) $p(X)$

- ex. 주변(marginal)확률 분포, 조건부(conditional)확률 분포

## 확률 변수의 함수
확률 변수 $X$의 함수 $Y = f(X)$도 확률 변수이다.

확률변수$X$의 함수  $Y = f(X)$의 역함수 $w(Y) = X$가 주어졌을 때 다음이 성립한다. 
> $p_y(y) = p_x(x)|\frac{dx}{dy}|$
- 즉, $x$에 대한 밀도함수에 $|\frac{dx}{dy}|$를 곱해 $y$의 밀도함수를 구할 수 있다.

---

### 결합 확률 밀도 함수(Joint PDF)
k차원의 확률변수 벡터 $\bold x = (x_1,x_2, \cdots, x_k)$가 주어지며, $k$개의 $\bold x$개의 함수들 $y_i = g_i(\bold x) (i = 1, \cdots, k)$를 정의하자. 즉 아래처럼 표현한다.
>$$\bold y = \bold g(\bold x)$$
$\bold y = \bold g(\bold x)$가 일대일 변환인 경우($\bold x = \bold w(\bold y)$가 유일해를 가짐), $\bold y$의 결합밀도함수를 아래와 같이 정의한다.
> $$p_y(y_1,\cdots,y_k) = p_x(x_1,\cdots,x_k)|\bold J|$$

즉, $|\bold J|$ 야코비얀 행렬을 곱한 형태.
> $$|\bold J|= \begin{vmatrix}
\frac{\partial x_1}{\partial y_1} & \cdots & \frac{\partial x_1}{\partial y_k} \\
\vdots & \ddots & \vdots\\
\frac{\partial x_k}{\partial y_1} & \cdots & \frac{\partial x_k}{\partial y_k}  
\end{vmatrix}$$

- 적용 예제 : Inverse CDF Technique
확률 변수 $X$가 CDF $F_X(x)$를 가진다고 하자. 연속확률분포함수 $U$ ~ $UNIF(0,1)$의 함수로 정의되는 다음 혹률 변수 $Y$를 고려하고자 한다.

> $$Y = F_X^{-1}(U)$$
>![Alt text](./images/image-12.png)
- CDF $F_X(x)$를 가지는 확률 변수 $X$에 대해 sampling을 하려고 할 때, 균등분포를 가지는 확률변수$U$에 대해 또다른 확률 변수$Y$를 정의하여 $Y$로부터 sampling을 할 수 있음을 보여준다.
- 확률 변수 $Y$는 확률변수 $X$와 동일한 분포를 따르게 된다.
>$$\begin{aligned}
F_y(y) =& P(Y ≤ y)\\
=& P(F_X^{-1}(U) ≤ y) \\
=& P(U ≤ F_x(y))\\
=& F_x(y)
\end{aligned}$$
![Alt text](./images/image-13.png)

- 추가 설명

CDF $F(x)$가 있고, 이의 역함수 $F^{-1}$에 $U$로부터 sampling하여 대입하면 $X$의 분포를 따르게 된다.

```python
# Inverse transform sampling
# CDF : F(d) = d**2 / r**2
# inverse CDF : r*(u**0.5)

import turtle, math, random

window = turtle.Screen()
turtle.tracer(8,0)
alex = turtle.Turtle()
alex.hideturtle()

r = 200
for i in range(5000):
    u = random.random() # 0~1 사이 : uniform distribution을 따름
    d = r * (u**0.5) 
    #d = u*r # d의 uniform한 샘플링 -> 진행되지 않음
    
    theta = random.random()*360
    x = d*math.cos(math.radians(theta))
    y = d*math.sin(math.radians(theta))

    alex.penup()
    alex.setposition(x,y)
    alex.dot()

turtle.update()
window.mainloop()
```

## 기댓값
- 기댓값 : 확률 분포 $p(x)$하에서 함수 $f(x)$의 평균값
- 이산확률분포 
>$E[f] \sum_xp(x)f(x)$ 
- 연속확률분포 
>$E[f] = \int p(x)f(x)dx$
- 확률 분포로부터 $N$개의 샘플을 추출해 기댓값을 근사할 수 있다.
> $$E[f] \approx \frac{1}{N}\sum_{n=1}^{N}f(x_n)$$

- 여러개 변수들의 함수
> $$E_{x}[f(x,y)] = \sum_{x}f(x,y)p(x)$$
즉, $x$에 대한 확률 밀도함수와 곱해 $x$에 대해 더한다.
연산 결과로 $y$에 대한 함수가 형성된다.

- $y$에 대한 함수 : 두개의 확률변수에 대한 합을 구한다.
> $$E_{x,y}[f(x,y)] = \sum_y\sum_{x}f(x,y)p(x,y)$$

- 조건부 기댓값
> $$E_x[f|y] = \sum_x f(x)(x|y)$$
> - $E_x[f|y]$ : $x$에 대한 확률밀도함수에 대해 $y$가 주어졌을 때의 조건부 확률에 대한 기댓값

## 분산과 공분산
- $f(x)$의 분산 : $f(x)$의 값들이 기댓값 $E[f]$으로부터 흩어져있는 정도.
>- $\begin{aligned}
var[f] =& E[(f(x) - E[f(x)])^2]
=& E[f(x)^2] - E[f(x)]^2
\end{aligned}$ 
>- $var(x) = E[x^2] - E[x]^2$
- 두개의 확률변수 $x,y$에 대한 공분산(Covariance)
>$$\begin{aligned}cov[x,y] =& E_{x,y}[\{x-E(x)\}\{y-E(y)\}]\\
=& E_{x,y}[xy] - E[x]E[y]
\end{aligned}$$
- $\bold x, \bold y$가 각각 확률변수의 벡터이다.
> $$\bold x = \begin{pmatrix}x_1\\x_2\\ \vdots \\ x_n \end{pmatrix}, \bold y = \begin{pmatrix}y_1\\y_2\\ \vdots \\ y_n \end{pmatrix}$$

>$$\begin{aligned}cov[\bold x,\bold y] =& E_{\bold x,\bold y}[\{\bold x-E(\bold x)\}\{\bold y^T-E(\bold y^T)\}]\\
=& E_{\bold x,\bold y}[\bold x\bold y^T] - E[\bold x]E[\bold y^T]
\end{aligned}$$
>$$cov[\bold x] \equiv cov[x,x] $$
- $xy^T$의 결과로 $n\times n$형태의 행렬을 얻으며, 이 행렬 원소의 평균값을 얻는다.

## 빈도주의 Vs 베이지안 (Frequentist VS Baysian)
확률을 해석하는 두가지 다른 관점 
- 빈도주의 : 반복 가능한 사건들의 빈도수에 기반
- 베이지안 : 불확실성을 정량적으로 표현

반복 가능하지 않은 사건일 경우, 이에 대한 불확실성을 정량적으로 표현 가능함.(북극 얼음이 세기말까지 녹아 없어질 확률!)
이는 우리가 **알고있는 정보**(얼음이 녹는 속도)에 근거해 확률을 정량적으로 나타낸다. 새로 수집하는 정보에 따라 확률을 업데이트 할 수 있다.

> $Q$. 모델의 파라미터 $\bold w$에 대한 우리의 지식을 확률적으로 나타내려면?

### Baysian
- $\bold w$에 대한 사전 지식 $p(\bold w) \to$ 사전확률(prior)
- 새로운 데이터 $D= {t_1, \cdots, t_N}$를 관찰하고 난 뒤의 조건부확률 $p(D|\bold w) \to$ 우도함수(likelihood function). 특정$\bold w$에 대해 $D$의 관찰값이 얼마나 가능성 있는지를 나타냄.
> $$p(\bold w|D) = \frac{p(D|\bold w)p(\bold w)}{p(D)}$$

- $p(\bold w|D)$는 $D$를 관찰하고 난 뒤의 $\bold w$불확실성을 표현한다.
> 사후확률 $\propto$ 우도 $\times$ 사전확률

 
### Freqeuntist
빈도주의는 $\bold w$가 고정된 파라미터이다. 최대 우도와 같은 `추정자(Estimator)`를 사용하여 그 값을 구한다. 구해진 파라미터의 불확실성은 `Bootstrap`을 사용하여 구할 수 있다.

### Baysian 장점
- 사전확률을 모델에 포함 가능
- 동전 던졌을 때 세번 다 앞면 나왔을 떄 
> - 빈도주의 : 최대 우도가 1이 됨(다음 동전던질 때 앞면 나올 확률을 1로 고려)
>- 베이지안 : 극단적 확률을 피할 수 있음. 사전지식(prior : $\frac{1}{2}$)가 포함되기 때문.

### 정규 분포 최대 우도해(maximum likelihood solution)
- 우도 : $\bold X = (x_1, \cdots, x_N)^T$가 독립적으로 같은 가우시안 분포로부터 추출된 N개의 샘플들이 가지는 확률
> $$\begin{aligned}p(\bold X : \mu,\sigma^2) =& p(x_1, \cdots, x_N | \mu,\sigma^2) = \prod_{n=1}^NN(x|\mu,\sigma^2)\\
\ln p(\bold X : \mu,\sigma^2) =& -\frac{1}{2\sigma^2}\sum_{n=1}^N(x_n-\mu)^2-\frac{N}{2}\ln\sigma^2 - \frac{N}{2}\ln(2\pi)
\end{aligned}$$

## 곡선 근사 (Curve Fitting) : 확률적 관점
- 학습 데이터
> $$\bold X = (x_1,\cdots, x_N)^T, \bold t = (t_1, \cdots, t_N)^T$$
- 목표값 $\bold t$의 불확실성을 다음과 같은 확률 분포로 나타낼 수 있다.
> $$p(t|x,\bold w, \beta) = N(t|y(x,\bold w), \beta^{-1}$$
- parameters
>$$\bold w, \beta$$

### 파라미터들의 최대우도 해를 구해보자!
- 우도 함수 
> $$p(t | x, \bold w, \beta) = \prod_{n=1}^NN(t_n|y(x_n,\bold w),\beta^{-1})$$
- 로그우도함수
> $$\ln p(t | x, \bold w, \beta) = -\frac{\beta}{2}\sum_{n=1}^{N}(y(x_n, \bold w)-t_n)^2 - \frac{N}{2}\ln\beta - \frac{N}{2}\ln(2\pi)$$

이 때 $\bold w$에 관해 **우도함수를 최대화** 시키는 것은 제곱합 **오차함수$-\frac{\beta}{2}\sum_{n=1}^{N}(y(x_n, \bold w)-t_n)^2$를 최소화** 시키는 것과 동일함을 알 수 있다.

![Alt text](./images/image-23.png)
- y값에 대한 불확실성을 고려하고자 $x_0가 주어졌을 때 $t$의 확률 $p(t | x, \bold w, \beta)$은 그림과 같이 가우시안 분포를 나타내고 있다고 가정한다.

- $\beta$의 최대우도해 
> $$\frac{1}{\beta_{ML}} = \frac{1}{N}\sum_{n=1}^N(y(x_n, \bold w_{ML})- t_n)^2$$

- 예측 분포
> $$p(t|x,\bold w_{ML}, \beta_{ML}) = N(t|y(x,\bold w_{ML}), \beta^{-1}_{ML})$$

#### 다음 단계 : 사전확률 모델에 포함하기
- 파라미터 $\bold w$의 사전확률
> $$p(\bold w|\alpha) = N(\bold w | 0, \alpha^{-1}\bold I) = (\frac{\alpha}{2\pi})^{\frac{M+I}{2}}e^{-\frac{\alpha}{2}\bold w^T\bold w}$$

- Baysian 정리에 따르면, 사후확률이 우도함수와 사전확률 곱에 비례한다.
> $$p(\bold w|\bold X, \bold t, \alpha, \beta) \propto p(t|\bold X, \bold w, \beta)p(\bold w|\alpha)$$
- 이 사후확률을 최대화하는 것이 아래를 최소화 시키는 것과 동일. (규제항이 포함되어있다.)
> $$\frac{\beta}{2}\sum_{n=1}^N(y(x_n, \bold w_)- t_n)^2 +\frac{\alpha}{2}\bold w^T\bold w$$
- 이는 규제화된 제곱합 오차함수를 최소화 시키는 것과 동일 ($\lambda = \frac{\alpha}{\beta}$)

#### 최종 단계 : 완전한 곡선 근사
위에서는 $t$의 예측 분포를 구하기 위해 $\bold w$의 점 추정에 의존했었다. 완전한 베이지안 방법은 $\bold w$의 분포로부터 확률의 기본 법칙만을 사용해 $t$의 예측 분포를 유도한다.
>$$p(t|x,\bold X, \bold t) = \int p(t|x,\bold w)p(\bold w|\bold X,\bold t)dw$$
- 식 안에 $\bold w$의 사후확률 $p(\bold w|\bold X,\bold t)$을 구해야하고, 여기서의 우도와 사전확률을 구하는 과정이 필요함.

>![Alt text](./images/image-24.png)
- 모델링을 하면, target값이 하나만 나오는게 아닌 확률 분포가 나오게 된다.
- 베이시안 방법으로 $t$의 예측 분포를 얻을 수 있다.
- 범위가 색으로 칠해진 것(밴드)처럼 예측 범위가 나오게 된다.
- 원래의 점들을 생성했던 함수가 밴드 내에 머물러 있음을 확인할 수 있다.

## 선형회귀 (Linear Regression)

먼저, 주어진 데이터를 직선을 사용해 모델링하는 방법을 살펴본다. 직선함수는 다음과 같은 형태를 가진다.
$$
y = ax + b
$$
여기서 $a$는 기울기 (*slope*)이고 $b$는 $y$절편 (*intercept*)라고 불린다.

아래 그래프는 기울기가 2이고 $y$절편이 -5인 직선으로부터 생성된 데이터를 보여준다

```python
rng = np.random.RandomState(1)
data_num = 1000
x = 10 * np.random.normal(0,1,data_num) # 정규분포를 따르는 x에서 sampling
y = 2 * x - 5 + rng.randn(data_num)
plt.scatter(x, y);
```

```python
rng = np.random.RandomState(1)
data_num = 1000
x = 10 * np.random.uniform(0,1,data_num) # 균등 분포를 따르는 x에서 sampling
y = 2 * x - 5 + rng.randn(data_num)
plt.scatter(x, y);
```

Scikit-Learn의 ``LinearRegression`` estimator를 사용해서 위 데이터를 가장 잘 표현하는 직선을 찾을 수 있다.

```python
x.shape, y.shape
```

```python
x[:50, np.newaxis].shape, y[:50].shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

rang = 50
model.fit(x[:, np.newaxis], y[:])

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit)
```

모델 학습이 끝난 후 학습된 파라미터들은 model."파라미터이름"_ 의 형태로 저장된다. 기울기와 y절편은 아래와 같이 출력할 수 있다.

```python
print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)
```

``LinearRegression`` estimator는 위의 예제와 같은 1차원 입력뿐만 아니라 다차원 입력을 사용한 선형모델을 다룰 수 있다. 다차원 선형모델은 다음과 같은 형태를 가진다.

$$
기하학적으로 이것은 hyper-plane(초평면)으로 데이터를 표현하는 것이라고 말할 수 있다.

## 선형 기저함수 모델 (Linear Basis Function Models)

비선형데이터를 선형함수로 모델링하는 한가지 방법은 기저함수(basis function)을 사용하는 것이다.

예를 들어, 다음과 같은 선형함수를 사용한다고 하자.
$$
y = a_0 + a_1 x_1 + a_2 x_2 + a_3 x_3 + \cdots
$$
여기서 $x_1, x_2, x_3,$등을 1차원 $x$로부터 생성할 수 있다($x_n = f_n(x)$). $f_n$을 기저함수(basis function)라고 부른다.

만약 $f_n(x) = x^n$라는 기저함수를 사용하면 최종적인 모델은 다음과 같을 것이다.

$$
y = a_0 + a_1 x + a_2 x^2 + a_3 x^3 + \cdots
$$

이 모델은 여전히 계수($a_n$)에 관해서는 선형함수임을 기억하자. 따라서 1차원 변수인 $x$를 기저함수를 통해 다차원으로 확장시킴으로써 우리는 여전히 선형모델(linear regression)을 사용할 수 있게 된다.

### 다항 기저함수 (Polynomial Basis Functions)

$f_n(x) = x^n$ 형태의 함수를 **다항 기저함수 (polynomial basis functions)** 라고 부른다. Scikit-Learn은 ``PolynomialFeatures``이라는 transformer를 이미 포함하고 있다.

```python
from sklearn.preprocessing import PolynomialFeatures
x = np.array([2, 3, 4])
poly = PolynomialFeatures(3, include_bias=False)
poly.fit_transform(x[:, None])
```

PolynomialFeatures가 1차원 array를 3차원 array로 변환한 것을 볼 수 있다. 이렇게 변환된 데이터를 선형모델에 적용할 수 있다.

7차원 변환을 적용해보자.

```python
# 여러 스텝의 일들을 한번에 처리
from sklearn.pipeline import make_pipeline
poly_model = make_pipeline(PolynomialFeatures(7),
                           LinearRegression())
```

다차원 변환을 사용하면 복잡한 데이터를 모델링할 수 있게 된다. 예를 들어 sine 함수를 사용해서 데이터를 생성하고 모델링해보자.

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(100)
y = np.sin(x) + 0.1 * rng.randn(100) # sin분포를 따르는 데이터 생성
plt.scatter(x, y)
```

```python
poly_model.fit(x[:, np.newaxis], y)
yfit = poly_model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);
```

### 가우시안 기저함수 (Gaussian Basis Functions)

다항 기저함수 외에 다른 기저함수를 사용해보자. 가우시안 기저함수는 다음과 같이 정의된다. (여러 $u$에 대한 기저함수 생성)

$$\exp \{-\frac{(x-u_j)^2}{2s^2}\}$$

$u_j$는 함수의 위치, $s$는 폭을 결정한다. 주어진 데이터를 여러 개의 가우시안 기저함수들의 합으로 표현하려고 시도할 수 있다.

![](images/05.06-gaussian-basis.png)

그래프 아래 회색부분은 다수의 가우시안 기저함수들을 보여주고 있다. 이 기저함수들의 합은 데이터를 가로지르는 매끈한 곡선이 된다.

가우시안 기저함수는 Scikit-Learn에 포함되어 있지 않지만 어렵지 않게 구현할 수 있다.

```python
from sklearn.base import BaseEstimator, TransformerMixin

class GaussianFeatures(BaseEstimator, TransformerMixin):
    """Uniformly spaced Gaussian features for one-dimensional input"""
    
    def __init__(self, N, width_factor=2.0):
        self.N = N
        self.width_factor = width_factor
    
    @staticmethod
    def _gauss_basis(x, y, width, axis=None):
        arg = (x - y) / width
        return np.exp(-0.5 * np.sum(arg ** 2, axis))
        
    def fit(self, X, y=None):
        # create N centers spread along the data range
        self.centers_ = np.linspace(X.min(), X.max(), self.N)
        self.width_ = self.width_factor * (self.centers_[1] - self.centers_[0])
        return self
        
    def transform(self, X):
        return self._gauss_basis(X[:, :, np.newaxis], self.centers_,
                                 self.width_, axis=1)
    
gauss_model = make_pipeline(GaussianFeatures(20),
                            LinearRegression())
gauss_model.fit(x[:, np.newaxis], y)
yfit = gauss_model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit)
plt.xlim(0, 10);
```

## 규제화 (Regularization)

기저함수를 사용함으로써 복잡한 데이터를 모델링할 수 있게 되었지만 조심하지 않는다면 over-fitting이라는 다른 심각한 문제를 만날 수 있다! 예를 들어, 너무 많은 개수의 가우시안 기저함수를 사용하게 되면 다음과 같이 될 수 있다.

```python
model = make_pipeline(GaussianFeatures(60),
                      LinearRegression())
model.fit(x[:, np.newaxis], y)

plt.scatter(x, y)
plt.plot(xfit, model.predict(xfit[:, np.newaxis]))

plt.xlim(0, 10)
plt.ylim(-1.5, 1.5);
```

이 예제에서는 60개의 기저함수가 사용되었는데 모델이 필요이상으로 flexible해져서 데이터가 없는 곳에서는 극단적인 값을 가지는 것을 볼 수 있다. 기저함수의 계수들은 다음과 같이 확인할 수 있다.

```python
def basis_plot(model, title=None):
    fig, ax = plt.subplots(2, sharex=True)
    model.fit(x[:, np.newaxis], y)
    ax[0].scatter(x, y)
    ax[0].plot(xfit, model.predict(xfit[:, np.newaxis]))
    ax[0].set(xlabel='x', ylabel='y', ylim=(-1.5, 1.5))
    
    if title:
        ax[0].set_title(title)

    ax[1].plot(model.steps[0][1].centers_,
               model.steps[1][1].coef_)
    ax[1].set(xlabel='basis location',
              ylabel='coefficient',
              xlim=(0, 10))
    
d = 60
model = make_pipeline(GaussianFeatures(d), LinearRegression())
basis_plot(model)
```

위 두번째 그래프는 각각의 가우시안 기저함수의 크기(계수값)을 보여주고 있다. Over-fitting이 일어나는 영역에서는 인접한 기저함수들의 값이 극단으로 가면서 서로 상쇄하는 현상이 일어난다. 따라서 큰 계수값에 대해 penalty를 부여해서 over-fitting을 어느 정도 극복할 수 있을 것이다. 이러한 penalty를 *regularization*이라 부른다.

### Ridge regression ($L_2$ Regularization)

가장 자주 쓰이는 형태의 regularization은 *ridge regression* ($L_2$ *regularization*)이고 다음과 같이 정의된다.
$$
P = \alpha\sum_{n=1}^N \theta_n^2
$$
여기서 $\alpha$는 regularization의 강도를 조절하는 파라미터이다. 이 형태의 regularization은 Scikit-Learn의 ``Ridge`` estimator에서 사용된다.

```python
from sklearn.linear_model import Ridge
model = make_pipeline(GaussianFeatures(d), Ridge(alpha=0.1))
basis_plot(model, title='Ridge Regression')
```

$\alpha$값이 0에 가까워질수록 일반적인 선형회귀모델이 되고 $\alpha$값이 무한대로 증가하면 데이터는 모델에 영향을 주지 않게 된다.

### Lasso regression ($L_1$ regularization)

또 하나의 자주 쓰이는 regularization방법은 계수들의 절대값의 합을 제한하는 것이다.
$$
P = \alpha\sum_{n=1}^N |\theta_n|
$$
뒤에서 자세히 다루겠지만 이 방법은 *sparse*한 모델을 생성하게 된다(많은 계수들이 0이 됨).

```python
from sklearn.linear_model import Lasso
model = make_pipeline(GaussianFeatures(d), Lasso(alpha=0.001))
basis_plot(model, title='Lasso Regression')
```

위에서 볼 수 있듯 대부분의 계수값들이 0이 된다. Ridge regression과 마찬가지로 $\alpha$값으로 regularization의 강도를 조절할 수 있다.
