git은 `local 저장소`와 `remote 저장소`가 존재한다.

- 로컬 저장소 생성
>`git init`

- git 저장소의 파일 상태 확인
>`git status`
- 커밋에 반영할 파일 추가
>`git add .` or `git add file.txt`

- 커밋 추가 : .Git(local) 저장소에 저장
>`git commit -m 'message'`

>- 메세지 작성 시 참고하기
    
    
- 커밋한 기록 확인
>`git log`

## git의 branch


- 현 branch 확인
>`git branch -v`
- branch 생성
>`git branch <branch_name>`
- branch 전환하기 : branch_name로의 전환
>`git checkout <branch_name>`
- 두 branch를 합치기 : 다른 branch의 코드상태를 현 branch에도 반영하기
>`git merge <another branch name>`
