- x축, y축 이름 넣기

`.xlabel('<name>')`
`.ylabel('<name>')` 

```python
x = np.arange(-10,10,0.01)
plt.xlabel("x value")
plt.ylabel("y value")
plt.plot(x, x**2)
plt.show()
```

- x,y 축의 값 범위 지정
`.axis([x_min,x_max,y_mix,y_max])`

```python
x = np.arange(-10,10,0.01)
plt.xlabel("x value")
plt.ylabel("y value")
plt.axis([-5,5,0, 25])
plt.plot(x, x**2)
plt.show()
```

- x,y축 표기 범위와 표기 간격 정하기

`.xticks(value : iterable variable)`

```python
x = np.arange(-10,10,0.01)
plt.xlabel("x value")
plt.ylabel("y value")
plt.axis([-5,5,0, 25])
plt.xticks([i for i in range(-5, 6,1)])
plt.yticks([i for i in range(0,25,3)])
plt.plot(x, x**2)
plt.show()
```

- 제목 지정

`.title('<name>')`

```python
x = np.arange(-10,10,0.01)
plt.xlabel("x value")
plt.ylabel("y value")
plt.axis([-5,5,0, 25])
plt.xticks([i for i in range(-5, 6,1)])
plt.yticks([i for i in range(0,25,3)])
plt.title("y = x^2 graph")

plt.plot(x, x**2, label="trand")
plt.legend()
plt.show()
```

```python
#심심해서 sin그래프
import numpy as np
import math
plt.plot([math.sin((i/18)*math.pi) for i in np.arange(0,36,1e-5)]) # 실제 Plotting
plt.show() # plt를 확인
```

## II. Matplotlib Case Study

- plot : 시계열 데이터에서 자주 사용 (x : time)
- scatter : 상관관계 유추 시.

## 꺾은 선 그래프 (Plot)
`.plot(x,y)`

```python
x = np.arange(20)
y = np.random.randint(0,20,20) # 0~20사이 난수 20번 
plt.plot(x,y)
# y축을 20까지 보이게 하고 싶다면? 
plt.axis([0,20,0,20])
# y축을 5단위로만 보이게 하기.
plt.yticks(np.arange(0,20,5))

plt.xticks(np.arange(0,20,5))
plt.show()
```

### 산점도(scatter plot)
 `.scatter(x,y)`

```python
plt.scatter(x,y)
plt.show()
```

### 박스 그림(box plot)
`.boxplot(y)`
- 수치형 데이터에 대한 정보(`Q1`, `Q2`, `Q3`, `min`, `max`)

```python
plt.boxplot((x, y))
plt.title("Box plot of x, y")
plt.show()
#아래부터 최솟값, 하위 25%, 중앙값, 상위 25%, 최댓값)
```

##  막대그래프(Bar plot)

`.bar(x,y)`
- 범주형 데이터의 '값'과 그 값의 크기를 직사각형으로 나타낸 그림.

```python
plt.bar(x, y)
plt.xticks(np.arange(0, 20, 1))
plt.show()
```

## 히스토그램
`.hist(x,y)`
- 도수분포를 직사각형의 막대 형태로 나타냈다. 
- 계급으로 나타내는 것이 특징: **범주형 데이터**로 구성 후 그림을 그림
- 범주형 데이터 : `0,1,2`가 아닌 `0~2`까지의 데이터로 구성한 것
- 막대 그래프와 다르게 두 막대 사이 거리를 떨어뜨리지 않음.

```python
#히스토그램

plt.hist(y, bins=np.arange(0,20,2)) # bins : 간격
plt.xticks(np.arange(0,20,2))
plt.show()
```

## 원형 그래프(Pie chart)
`.pie(x, label)`
- 데이터 전체에 대한 부분의 비율을 부채꼴로 나타냄
- **비율** 확인에 용이


```python
z = [100,200,300,400]
plt.pie(z,labels=['one', 'two', 'three', 'four'])
plt.show()
```

## IV. The 멋진 그래프, Seaborn Case Study
- 커널밀도그림 (Kernel Density Plot)
- 카운트그림 (Count Plot)
- 캣그림 (Cat Plot)
- 스트립그림 (Strip Plot)
- 히트맵 (Heatmap)

```python
import seaborn as sns
```

```python
x = np.arange(0,22,2)
y = np.random.randint(0,20,20)
plt.hist(y, bins=x)
plt.show()
```

### 커널 밀도 그림 (kdeplot)
- 히스토그램과 같은 **연속적**인 분포를 곡선화하여 그린 plot
- `sns.kdeplot(y : 도수)`

```python
# kdeplot()
sns.kdeplot(y, shade=True) # shade : 음영색 넣기
plt.show()
```

### 카운트 그림 (Count plot)
- 범주형 column의 빈도수를 시각화 -> Groupby 후의 도수를 하는 것과 동일한 효과
- `sns.countplot()`

```python
#카운트그림 
vote_df = pd.DataFrame({"name":['Andy', 'Bob', 'Cat'],"vote":[True, True, False]})
vote_df
```

```python
vote_count = vote_df.groupby('vote').count()
vote_count
```

```python
plt.bar(x=[False,True],height=vote_count['name'])
plt.show()
```

```python
vote_df['vote']
```

```python
sns.countplot(x = vote_df['vote'])
plt.show()
```

### 캣그림(Cat plot)
- 숫자형 변수와 하나 이상의 범주형 변수 관계를 보여주는 함수
- `sns.catplot()`

```python
#catplot 숫자나 범주형 변수 관계
#s = sns.catplot(x=, y=, hue=, data=, kind="도표 형태")
#s.fig.set_size_inches(10,6)

covid = pd.read_csv('../3w/archive/country_wise_latest.csv')
covid.head()
```

```python
sns.catplot(x = 'WHO Region', y = 'Confirmed',data=covid, kind= 'violin') # kind : 'violin', 'strip'
plt.xticks(rotation = 45)
plt.show()
```

### 스트립그림 (strip Plot)
- scatter plot과 유사하게 데이터의 수치를 표현하는 그래프
- `sns.stripplot()`

```python
#strip plot
sns.stripplot(x='WHO Region', y = 'Confirmed', data=covid)
```

```python
#뭉친 데이터 분산하여 표현
sns.swarmplot(x='WHO Region', y = 'Confirmed', data=covid) 
```

### 히트맵(Heatmap)
- 데이터의 행렬을 색상으로 표현해주는 그래프
- `sns.heatmap()`

```python
# 상관계수
covid_corr=covid.drop(['Country/Region','WHO Region'], axis = 1).corr()
```

```python
# heatmap # 색을 통해 데이터를 표현
sns.heatmap(covid_corr)
plt.show()
```
