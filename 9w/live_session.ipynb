# AI model과 Serving

- 기본 개념
> - AI model : 특정 테스크 수행을 위해 명시적 프로그래밍 외에 예측이나 결정을 내릴 . 수있게 데이터에 대해 학습된 COmputational Structure
> - Model : 훈련된 모델이 서버나 데스크탑에 저장되어만 있는 것은 소용이 없다.
> - 업무 투입을 위해 모델의 학습단계와 서비스 제공단계를 구분하여 생각해야 함.

## 서빙 특성
AI와 실제 애플리케이션의 교차점
- 학습된 모델은 이전에 보지 못한 새로운 데이터에 대한 예측 혹은 의사결정을 하는데 사용된다.
- 서빙은 모델을 텍스트, 이미지, 사운드같은 입력을 받아 예측이나 의사결정에 필요한 형태의 출력을 생성할 . 수있도록 지속적으로 실행이 되면서 서비스와 product에 결과를 전달함.

## 배포
-> 주로 데브옵스 혹은 백엔드 엔지니어들의 작업
학습된 모델을 운영 환경에 통합하여 최종 사용자나 다른 시스템에서 사용할 수 있도록 하는 일련의 작업들
- 모델 서빙을 위한 인프라 환경 마련
- 여러 요청을 처리하기위한 scaling
- Robust & Fault-tolerant 모델을 서빙하고 운영

모델 배포 Architecture
> Usecase 또는 서비스 요구사항에 따라 배포방법은 다양하다.
온프레미스 또는 클라우드 서버에서 호스팅, 혹은 애플리케이션 자체에 내장하기도 함.

## 이해관계자
인공지능 프로젝트를 진행할 때 다양한 이해관계가 있는 개인이나 그룹

다양한 이해관계자들의 이해와 요구사항을 반영해야 함.



### 주요 관심 및 요구사항

투자수익률(ROI) 증대
- 작업 자동화와 인사이트 생성은 인공지능 개발에 따른 결과일 뿐, 이를 통한 수익 창출에 관심에 많다.
- 인공지능 서비스 적용에 따른 운영 비용 절감 등에 따른 비용 감소 등의 경제적 요소를 중요시

시장 내 기술 경쟁우위 확보
- 인공지능 서비스 개발과 배포에 따른 시장 내 경쟁우위 확보는 기업의 시장가치를 높임
- 실제로 챗gpt 등장에 따라 인공지능 기업 및 관련 기업들의 주식시장에서 가치가 크게 상승한 사례 존재

기업 리스크, 윤리 및 평판 관리
- 대중의 인식이나 인공지능 서비스 도입 시 발생할 수 있는 리스크나 대중의 인식에 대해 고민 필요
- 단순한 경제적 이유로 프로젝트 진행 않을 수도 있고, 모델 제공에 따른 사회적 역할과 책임에 대한 논란 발생 시, 모델 자체가 흔들릴 수 있다.

인공지능 모델의 성능 일관성, 강건성
- 모델의 출력 결과가 일관된 성능을 보여야 안정적 서비스 운영이 가능하다.
- 서빙과 배포 후 다른 시스템 구성요소와 원활한 작동이 되어야 함.
- 모델 업데이트와 반복적인 개선을 효율적으로 처리하여 모델의 성능 일관성, 강건성을 확보해야 함.

시장 니즈와 구현 가능성의 접점 찾기
- 모델 자체가 가진 구조적 한계와 제약 사항을 명확히 하기.
- 이런 제약을 인식한 후, 서비스 레벨에서 사용자의 경험적인 요소로 이를 극복하는 것 고려
- 올바른 벤치마크 설정 후 모델의 현실적 기대치 조정하기

**기술 부채에 대한 대비**
- 모든 기술은 수명주기가 존재함. 인공지능은 그 주기가 더욱 짧고 빠르게 변화함.
- 가능한 Techinal Debt를 줄일 수 있는 모델 아키텍처와 서빙, 배포 프레임워크를 선택해야함.
- 성급한 배포는 유지보수 비용을 증가시키고 잠재적 장애 발생 여지를 남김.

## 사용자 측면의 주요 관심 및 요구사항
인공지능 모델, 솔루션의 신뢰성과 유연성
- 인공지능 모델의 성능이 일관되지 못해 서비스 사용에 대한 신뢰성이 떨어지게 됨. 
- OpenAI의 GPT API가 초기에넌 버전이나 모델의 성능이나 버전의 변화에 대해 투명하게 운영되지 못해 사용률 저조한 시기가 존재했음
- 모델의 현실적인 제약이 있음에도 이를 서비스 레벨에서 정확하게 작동한다는 인지를 주지 못하면 서비스 자체의 신뢰성이 손상됨.

사용자의 경험
- 잘못된 모델 서빙, 배포 전략으로 응답시간이 길어지면 사용자가 도망감

## 서빙 시 내부 동작 프로세스

Model Loading
- 사전에 학습된 모델을 메모리를 불러와 서빙을 예측하도록 하는 작업
- 모델의 크기가 클수록 모델 로딩 과정에서 많은 자원과 리소스를 필요로 할 수 있다.
- 하드웨어에 따라 Deserialization 방법이 다를 수 있다.

Inference
- CPU, GPU 메모리에 올라간 모델에 요청된 입력을 넣어 예측을 수행하는 작업 
- 모델 로딩 이후의 서빙 측면에서는 추론과정에서 연산 비용이 커 이 부분에 대한 최적화 작업을 많이 연구함.
- Hardware-Oriented 최적화 작업을 요구하여 쉽지 않으나 다양한 프레임워크에서 이를 업데이트 및 지원 

Response Handling
- 모델에서 예측결과는 벡터 혹은 스칼라와 같은 수치젹 결과로 출력, 이를 사용자에게 친숙한 출력으로 전달하기 위한 핸들링 필요
- 요청에 대한 결과 반환까지의 전체 작업을 포함하고 있어 서빙 측면에서는 입출력 데이터의 전후처리에 대한 최적화와 네트워크 통신 방법 등에 대한 여러가지 작업을 동반함.

## 실시간 서빙 Vs 배치 서빙
모든 서비스가 사용자에게 실시간으로 결과를 전달할 필요는 없음.
최종 사용자가 서비스 사용자 혹은 고객인 경우에 모델의 예측 결과가 실시간으로 제공되어야 할 수는 있음.
하지만 비즈니스 로직 구현 방식에 따라 한번에 배치로 예측을 미리 해둘 수 있다.

# Container
컨테이너 기반의 모델 서빙과 배포의 중요성
도커파일정도의 설계는 필요함.

- Consistency
>- 일정범위에서 환경이 변화해도 동일한 서빙동작과 결과를 제공할 수 있다.
>- 일관된 서비스를 제공할 수 있다.

- Isolation
>- 다른 애플리케이션의 간섭 없이 컨테이너가 하나의 격리된 상태로 별도 실행 가능
>- 서빙 시 여러 시스템 간의 충돌등으로 일어나는 이슈나 문제가 감소함

- Portability
>- 다양한 시스템에서 쉽게 컨테이너 이미지를 고유, 배포할 수 있음.
>- 비공개 private 이미지 repo도 사용할 수 있어 내부적으로 공유 배포 가능.
